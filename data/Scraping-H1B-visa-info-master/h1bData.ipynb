```python
from h1b_scraper import H1B_Scraper
```

```python
scraper = H1B_Scraper()
ds_jobs = ["Junior Data Scientist",
           "Associate Data Scientist",
           "Data Scientist",
           "Staff Data Scientist",
           "Senior Data Scientist", 
           "Principal Data Scientist",
           "Lead Data Scientist",
           "Data Scientist 1", 
           "Data Scientist 2", 
           "Data Scientist 3", 
           "Data Scientist I",
           "Data Scientist II",
           "Data Scientist III"
          ]
se_jobs = ["Junior Software Engineer",
           "Associate Software Engineer",
           "Software Engineer", 
           "Senior Software Engineer",
           "Staff Software Engineer",
           "Principal Software Engineer",
           "Lead Software Engineer",
           "Software Engineer 1", 
           "Software Engineer 2", 
           "Software Engineer 3",
           "Software Engineer I",
           "Software Engineer II",
           "Software Engineer III"
           ]
da_jobs = ["Junior Data Analyst",
           "Associate Data Analyst",
           "Data Analyst",
           "Staff Data Analyst",
           "Senior Data Analyst", 
           "Principal Data Analyst",
           "Lead Data Analyst",
           "Data Analyst 1", 
           "Data Analyst 2", 
           "Data Analyst 3",
           "Data Analyst I", 
           "Data Analyst II", 
           "Data Analyst III"
            ]
jobs = [] + da_jobs + ds_jobs + se_jobs
years = [year for year in range(2015, 2021)] ## get the list [2017, 2018]
tables = scraper.scrape(years, jobs)
```

```python
tables
```

```python
se_tables = []
for i in years:
    for j in se_jobs:
        table = scraper.extract_specific_table(tables, i, j)
        se_tables.append(table)
```

```python
da_tables = []
for i in years:
    for j in da_jobs:
        table = scraper.extract_specific_table(tables, i, j)
        da_tables.append(table)
```

```python
ds_tables = []
for i in years:
    for j in ds_jobs:
        table = scraper.extract_specific_table(tables, i, j)
        ds_tables.append(table)
```

```python
import pandas as pd
```

```python
da = pd.concat(da_tables)
ds = pd.concat(ds_tables)
se = pd.concat(se_tables)
```

```python
da.columns = ['employer', 'title', 'salary', 'location', 'submitDate', 'startDate', 'caseStatus']
ds.columns = ['employer', 'title', 'salary', 'location', 'submitDate', 'startDate', 'caseStatus']
se.columns = ['employer', 'title', 'salary', 'location', 'submitDate', 'startDate', 'caseStatus']
```

```python
ds.location.str.split(", " ,expand=True)
```

```python
da[['city', 'state', 'none']] = da.location.str.split(", " ,expand=True)
ds[['city', 'state', 'none']] = ds.location.str.split(", " ,expand=True)
se[['city', 'state', 'none1', 'none2']] = se.location.str.split(", ", expand=True)
```

```python
da.drop(['none'], axis=1, inplace=True)
ds.drop(['none'], axis=1, inplace=True)
se.drop(['none1', 'none2'], axis=1, inplace=True)
```

```python
da['employer'] = da['employer'].astype(str)
da['title'] = da['title'].astype(str)
da['location'] = da['location'].astype(str)
da['caseStatus'] = da['caseStatus'].astype(str)
da['submitDate'] = da['submitDate'].astype(str)
da['startDate'] = da['startDate'].astype(str)

ds['employer'] = ds['employer'].astype(str)
ds['title'] = ds['title'].astype(str)
ds['location'] = ds['location'].astype(str)
ds['caseStatus'] = ds['caseStatus'].astype(str)
ds['submitDate'] = ds['submitDate'].astype(str)
ds['startDate'] = ds['startDate'].astype(str)

se['employer'] = se['employer'].astype(str)
se['title'] = se['title'].astype(str)
se['location'] = se['location'].astype(str)
se['caseStatus'] = se['caseStatus'].astype(str)
se['submitDate'] = se['submitDate'].astype(str)
se['startDate'] = se['startDate'].astype(str)
```

```python
da['submitDate'] = pd.to_datetime(da['submitDate'])
da['startDate'] = pd.to_datetime(da['startDate'])

ds['submitDate'] = pd.to_datetime(ds['submitDate'])
ds['startDate'] = pd.to_datetime(ds['startDate'])

se['submitDate'] = pd.to_datetime(se['submitDate'])
se['startDate'] = pd.to_datetime(se['startDate'])
```

```python
da.drop(['location'], axis=1, inplace=True)
ds.drop(['location'], axis=1, inplace=True)
se.drop(['location'], axis=1, inplace=True)
```

```python
da.to_csv("../h1bDataAnalyst.csv")
ds.to_csv("../h1bDataScientist.csv")
se.to_csv("../h1bSoftwareEngineer.csv")
```
