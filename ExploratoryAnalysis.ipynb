```python
import pandas as pd
import numpy as np
from datetime import datetime
from sklearn import linear_model as lm
import matplotlib.pyplot as plt
```

```python
bwData = pd.read_excel("data/jobGrowthSalariesByYear.xlsx")
```

```python
bwData.columns = ['year', 'min', 'fq', 'sq', 'tq',
       'max']
bwData
```

```python
fig, ax = plt.subplots(figsize=(16, 8))
ax.plot(bwData.year, bwData.fq, marker='o', label='First Quartile')
ax.plot(bwData.year, bwData.sq, marker='o', label='Second Quartile')
ax.plot(bwData.year, bwData.tq, marker='o', label='Third Quartile')
ax.set_xticklabels([0] + list(bwData["year"]))
plt.title("Tracking Salary Quantiles", fontsize = 18)
plt.xlabel("Year", fontsize=16)
plt.ylabel("Salary", fontsize=16)
ax.legend();
```
